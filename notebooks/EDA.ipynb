```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pathlib import Path

plt.style.use("default")
DATA_DIR = Path("../data/processed")
train = pd.read_csv(DATA_DIR / "train.csv")
test  = pd.read_csv(DATA_DIR / "test.csv")

train.shape, test.shape
```

```python
train['Class'].value_counts()
```

```python
fig , ax = plt.subplots(1,2,figsize=(12, 6))

train['Class'].value_counts(normalize=True)
sns.countplot(x='Class', data=train, ax=ax[0])
ax[0].set_title('Fraud (1) vs Non-Fraud (0) countplot', fontsize=14)

counts = train['Class'].value_counts(normalize=True)*100
counts.plot(kind='pie',autopct='%.1f%%',ax=ax[1])
ax[1].set_title('Fraud (1) vs Non-Fraud (0) pieplot', fontsize=14)

plt.tight_layout()
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,4))
sns.histplot(x=train['Amount'], bins=80, ax=axs[0])
axs[0].set_title("Amount distribution")
sns.histplot(x=train['Time'], bins=80, ax=axs[1])
axs[1].set_title("Time distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=train)
plt.yscale('log')
plt.title("Amount by Class (log scale)")
plt.show()
```

```python
cols = ['V1','V2','V3','V4','V5','V6','V7','V8','V9','V10','Amount','Time','Class']
corr = train[cols].corr()
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation (subset)")
plt.show()
```
